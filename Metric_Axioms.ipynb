```python
###Third Axiom###
import os
import pandas as pd
import numpy as np

df_RMSD = pd.read_csv(r'C:\Users\Clamfighter\Machine_Learning_Project\my_env\Masters\Output\JS_CrystalComparisonRMSD(full).csv', index_col=[0])
df_output = pd.DataFrame(columns=['crystal_1','crystal_2','crystal_3','RMSD_12','RMSD_23','RMSD_13','triangle_Defect'])
crystals = df_RMSD.columns

#for each triple of unique crystals (no doubles/triples)
for crystal_x in crystals:
    for crystal_y in crystals:
        if crystal_y is not crystal_x:
            RMSD_xy = df_RMSD[crystal_x][crystal_y]
            for crystal_z in crystals:
                if (crystal_z != crystal_x) and (crystal_z != crystal_y):
                    #get RMSD values between the 3 crystals
                    RMSD_yz = df_RMSD[crystal_y][crystal_z]
                    RMSD_xz = df_RMSD[crystal_x][crystal_z]
                    RMSD_values = [RMSD_xy,RMSD_yz,RMSD_xz]
                    #find the max of the RMSD values
                    maxRMSD = RMSD_values.index(max(RMSD_values))
                    #reorder the crystals and RMSDs to put the max first
                    #this let's us do the triangle defect calculation, and save to a file in a way we can understand later
                    if maxRMSD ==0:
                        crystal_order = [crystal_x,crystal_y,crystal_z]
                        RMSD_order = RMSD_values
                    elif maxRMSD == 1:
                        crystal_order = [crystal_y,crystal_z,crystal_x]
                        RMSD_order = [RMSD_yz,RMSD_xz,RMSD_xy]
                    elif maxRMSD == 2:
                        crystal_order = [crystal_z,crystal_x,crystal_y]
                        RMSD_order = [RMSD_xz,RMSD_xy,RMSD_yz]
                    #Calculate Triangle Defect
                    triangle_defect = RMSD_order[0]-RMSD_order[1]-RMSD_order[2]
                    row = crystal_order + RMSD_order 
                    row.append(triangle_defect)
                    df_output.loc[len(df_output)] = row
df_output #dataframe that is saved to CSV
```
